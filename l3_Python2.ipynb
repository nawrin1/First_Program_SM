```python
#H1 Write a function to calculate the sum of numbers from 1 to N.
def summation(num):
  total=0
  for i in range(1, num+1):
    total=total+i

  return total
print(summation(int(input("Enter number: "))))
```

```python
#H2 Write a function that returns the square of a number.
def square(num):
  return num**2
print(square(int(input("Enter number: "))))
```

```python
#H3 Write a function that checks whether a number is even/odd.
def even_odd(num):
  if num%2==0:
    return "The number is Even"
  else:
    return "The number is Odd"
print(even_odd(int(input("Enter number: "))))
```

```python
#H4 Write a function that returns the larger number from two given numbers.
def largerNumber(num1, num2):
  if num1>num2:
    return num1
  else:
    return num2
print(largerNumber(int(input("Enter number: ")), int(input("Enter number: "))))
```

```python
#H5  	Write a Python program to create a list of 10 integers. Then perform the following operations:
  #Append an element to the list.
  #Insert an element at a specific position.
  #Remove an element from the list.
  #Pop an element from a specific position.
list1=[]
def createList(num):

  for i in range(1, num+1):
    list1.append(i)
  return list1
print(createList(10))

#append
def appendList(num):
  list1.append(num)
  return list1

print(appendList(int(input("Enter number to append: "))))

def insertElement(num, pos):
  list1.insert(pos, num)
  return list1
print(insertElement(int(input("Enter number to insert: ")), int(input("Enter position: "))))

def removeElement(num):
  list1.remove(num)
  return list1

print(removeElement(int(input("Enter number to remove: "))))

def popElement(pos):
  if pos < len(list1):
        list1.pop(pos)
  else:
        print("Invalid position")
  return list1

print(popElement(int(input("Enter position to pop element: "))))
```

```python
#H6 Write a program to find all unique elements in the list.
list1=[1,2,2,4,5,5,7,8,9,10]
def uniqueElements(list1):
  unique=[]
  for i in list1:
    if i not in unique:
      unique.append(i)
  return unique
print(uniqueElements(list1))
```

```python
#H7 Write a Python program to calculate the mean of a list of numbers.
list1=[1,2,3,6,7]
def mean(list1):
  sum=0
  for i in range(len(list1)):
    sum=sum+list1[i]
  return sum/len(list1)


print(mean(list1))
```

```python
#H8 Classify Numbers into Even and Odd Lists
list1=[1,3,4,5,6,7,11,23,45]
even=[]
odd=[]
def evenOdd(list1):

  for i in list1:
    if i%2==0:
      even.append(i)
    else:
      odd.append(i)

evenOdd(list1)
print("Even List: ", even)
print("Odd List: ", odd)
```
