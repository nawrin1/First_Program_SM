Try these on your own — solutions follow the same patterns you just saw above.

**Matplotlib**
1. Bar chart of membership types — `gym['abonoment_type'].value_counts()`
2. Histogram of `avg_time_in_gym`
3. Scatter plot of `Age` vs `visit_per_week`

**Seaborn**
1. Box plot — `gender` vs `Age`
2. Count plot — `days_per_week`
3. Bar plot — mean `visit_per_week` per `abonoment_type`

```python

# Bar chart of membership types
membership_counts = gym['abonoment_type'].value_counts()

fig, ax = plt.subplots()
ax.bar(membership_counts.index, membership_counts.values, color=['#4caf50', '#ff9800', '#f44336'])
ax.set(title='Membership Types', xlabel='Abonoment Type', ylabel='Count')
plt.show()

```

```python
# Histogram of avg_time_in_gym
fig, ax = plt.subplots()
ax.hist(gym['avg_time_in_gym'], bins=20, color='purple', edgecolor='white')
ax.set(title='Average Time in Gym', xlabel='Minutes', ylabel='Frequency')
plt.show()
```

```python


# 3. Scatter plot of Age vs visit_per_week
fig, ax = plt.subplots()
ax.scatter(gym['Age'], gym['visit_per_week'], alpha=0.6, color='teal')
ax.set(title='Age vs Visits per Week', xlabel='Age', ylabel='Visits per Week')
plt.show()
```

```python
# Box plot :gender vs Age
import seaborn as sns

plt.figure()
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Age Distribution by Gender')
plt.show()
```

```python

# Count plot — days_per_week
plt.figure()
sns.countplot(data=gym, x='days_per_week', palette='viridis')
plt.title('Days per Week')
plt.show()

```

```python
# Bar plot — mean visit_per_week per abonoment_type
plt.figure()
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week', palette='pastel')
plt.title('Mean Visits per Week by Abonoment Type')
plt.show()
```
