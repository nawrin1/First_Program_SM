### Exercise 1 — NumPy
Given `arr = np.random.randint(1, 101, 20)`:
1. Print its sum, min, and max.
2. Create a second array of the same size and multiply them element-wise.

```python
# Your solution here
import numpy as np

# Create the array
arr = np.random.randint(1, 101, 20)

print("Array:", arr)
print("Sum:", np.sum(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))

#second array
arr2 = np.random.randint(1, 101, 20)
print("Second Array:", arr2)

#  multiplication
result = arr * arr2
print("Element-wise multiplication:", result)

```

### Exercise 2 — Pandas
Get a recent unique dataset from Kaggle or another source and compute basic statistics using Pandas

1. Choose a dataset — Download a current dataset from Kaggle
2. Load the dataset into a Pandas DataFrame.
3. Explore and inspect the data with Pandas methods.
4. Compute and display basic statistical summaries, including:
Descriptive statistics (describe)
Missing value counts
Data types
Class balance (if applicable)

```python
# Your solution here
import pandas as pd

# Load dataset
df = pd.read_csv("/content/clean_data.csv")

print(df.head())

```

```python
# Shape of dataset
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

df.info()

df.tail()
```

```python
# summary
print(df.describe())
```

```python
# Count missing values
print(df.isnull().sum())

# Percentage of missing val
print("\nPerentage of missing val:",(df.isnull().sum() / len(df)) * 100, sep="\n")
```

```python
print(df.dtypes)
```

```python
print(df['type'].value_counts())

# normalized
print(df['type'].value_counts(normalize=True).round(3) * 100)
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
#numericl distribution
print(df['price'].describe())
```

```python
#categorical distribution
print(df['transmission'].value_counts())
```
